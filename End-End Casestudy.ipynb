```python

df['order_purchase_timestamp'] = pd.to_datetime(df['order_purchase_timestamp'])

df['order_month'] = df['order_purchase_timestamp'].dt.to_period('M')

customers_per_month = df.groupby('order_month')['customer_unique_id'].nunique()
customers_per_month 
```

```python
#c. Understand the retention of customers on month on month basis
```

```python
retention_rates = []


for i in range(1, len(customers_per_month)):
    
    current_month_customers = customers_per_month.iloc[i]
    previous_month_customers = customers_per_month.iloc[i - 1]
    
    retained_customers = current_month_customers & previous_month_customers
     
    if (previous_month_customers.size)> 0:
        retention_rate = (retained_customers.size) / (previous_month_customers.size) * 100
    else:
        retention_rate = 0
    
    retention_rates.append((customers_per_month.index[i], retention_rate))

retention_df = pd.DataFrame(retention_rates, columns=['month', 'retention_rate'])
```

```python
#3. Cross-Selling (Which products are selling together)
#Hint: We need to find which of the top 10 combinations of products are selling together in each transaction.
#(combination of 2 or 3 buying together)
```

```python
grouped_data = df.groupby('order_id')['product_id'].apply(list)
```

```python
def generate_combinations(product_list, size):
    combinations = []
    length = len(product_list)
    for i in range(length):
        for j in range(i + 1, length):
            if size == 2:
                combinations.append((product_list[i], product_list[j]))
            elif size == 3:
                for k in range(j + 1, length):
                    combinations.append((product_list[i], product_list[j], product_list[k]))
    return combinations


combinations_2 = grouped_data.apply(lambda products: generate_combinations(products, 2))
```

```python
all_combinations_2 = [combo for sublist in combinations_2 for combo in sublist]
combination_counts_2 = pd.Series(all_combinations_2).value_counts()
```
